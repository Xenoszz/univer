```python
def distance(seq1, seq2):
  """
  ฟังก์ชันนี้คำนวณระยะทาง Hamming ระหว่างสองลำดับดีเอ็นเอ

  Args:
      seq1 (str): ลำดับดีเอ็นเอตัวแรก
      seq2 (str): ลำดับดีเอ็นเอตัวที่สอง

  Returns:
      int: ระยะทาง Hamming ระหว่างสองลำดับ
  """
  # เปรียบเทียบแต่ละตำแหน่งของลำดับ
  diff = 0
  for i in range(len(seq1)):
    if seq1[i] != seq2[i]:
      diff += 1
  return diff

def main():
  # รายการลำดับดีเอ็นเอ
  sequences = ["TGCTACTCCACACAC", "TAGTACTCCACACAC", "TGCTACTGCAGACAC", "TACTACTGCACCCAC", "TAGCACTCCAGACCC"]

  # คำนวณระยะทางระหว่างทุกคู่ของลำดับ
  distances = []
  for i in range(len(sequences)):
    for j in range(i + 1, len(sequences)):
      dist = distance(sequences[i], sequences[j])
      distances.append((i, j, dist))

  # แสดงระยะทางทั้งหมด
  for i, j, dist in distances:
    print(f"ระยะทางระหว่าง {sequences[i]} และ {sequences[j]}: {dist}")

if __name__ == "__main__":
  main()
```

```python
def sieve_of_eratosthenes(n):
    """
    Finds all prime numbers up to n using the Sieve of Eratosthenes algorithm.
    """
    is_prime = [True] * (n + 1)  # สร้างตารางและกำหนดให้ทุกตัวเลขเป็นเลขเฉพาะ
    is_prime[0] = is_prime[1] = False  # 0 และ 1 ไม่ใช่เลขเฉพาะ

    p = 2  # เริ่มจากเลขเฉพาะตัวแรก
    while p * p <= n:  # ทำซ้ำจนถึง √n
        if is_prime[p]:  # ถ้า p เป็นเลขเฉพาะ
            for i in range(p * p, n + 1, p):  # กรองพหุคูณของ p ออก
                is_prime[i] = False
        p += 1  # เลื่อนไปยังเลขเฉพาะถัดไป

    primes = [p for p in range(2, n + 1) if is_prime[p]]  # เก็บเลขเฉพาะทั้งหมด
    return primes

# ตัวอย่างการใช้งาน
n = 100
primes = sieve_of_eratosthenes(n)
print(f"Prime numbers up to {n}: {primes}")
```

```python
from collections import defaultdict

def longest_path(graph, start, end):
    # สร้าง dictionary เพื่อเก็บระยะเวลาสะสมไปยังแต่ละกิจกรรม และกิจกรรมก่อนหน้าในเส้นทาง
    dist = defaultdict(int)  # ระยะเวลาสะสมเริ่มต้นที่ 0 สำหรับกิจกรรมเริ่มต้น
    prev = defaultdict(list)  # กิจกรรมก่อนหน้า (อาจมีมากกว่า 1 ในกรณีมีเส้นทางวิกฤติหลายเส้น)

    # เริ่มต้นคิวสำหรับการสำรวจกิจกรรมแบบ breadth-first search (BFS)
    queue = [start] 
    for node in graph: 
        if node != start: 
            dist[node] = float('-inf')  # เริ่มต้นด้วยระยะอนันต์

    # สำรวจกราฟ
    while queue:
        curr = queue.pop(0)  
        for neighbor in graph[curr]:  # สำรวจข้าง
            duration = graph[curr][neighbor] 
            new_dist = dist[curr] + duration  # คำนวณระยะเวลาใหม่

            # อัปเดตระยะเวลาสะสม
            if new_dist > dist[neighbor]:  
                dist[neighbor] = new_dist
                prev[neighbor] = [curr] 
            elif new_dist == dist[neighbor]:
                prev[neighbor].append(curr)  
            if neighbor not in queue:  
                queue.append(neighbor)

    # สร้างเส้นทาง โดยย้อนกลับจากจุดสิ้นสุด
    critical_path = []
    curr = end
    while curr != start:
        critical_path.append(curr)
        curr = prev[curr][0]  
    critical_path.append(start)
    critical_path.reverse()  
    return dist[end], critical_path


# ตัวอย่างแผนภาพกิจกรรม
graph = {'START': {'A': 3, 'B': 4}, 'A': {'C': 2, 'D': 1}, 'B': {'E': 2}, 'C': {'F': 1}, 'D': {'FINISH': 0}, 'E': {'G': 1}, 'F': {'FINISH': 0}, 'G': {'FINISH': 0}, 'FINISH': {}}

# คำนวณเส้นทางวิกฤติ
critical_path_duration, critical_path = longest_path(graph, 'START', 'FINISH')
print(f"ระยะเวลาของเส้นทางวิกฤติ: {critical_path_duration} สัปดาห์")
print("เส้นทางวิกฤติ:", " -> ".join(critical_path))
```

```python
prices = [99, 2, 15, 17, 6, 35, 64, 10, 28, 32, 45, 11, 2, 36, 35, 44, 66, 9, 18, 24, 25, 98, 97, 11, 18, 2, 9, 19, 56, 12]

n = len(prices)
dp = [[[0, 0, 0] for _ in range(3)] for _ in range(n)]

for i in range(1, n):
    for j in range(1, 3):
        max_profit = float('-inf')
        for k in range(i):
            profit = dp[k][j-1][0] + max(0, prices[i-1] - prices[k])
            if profit > max_profit:
                max_profit = profit
                dp[i][j][0] = profit
                dp[i][j][1] = k  # store the buying day
                dp[i][j][2] = prices[k]  # store the buying price

max_profit = dp[-1][2][0]
buy_sell_days = []
for i in range(2, -1, -1):
    if dp[-1][i][0] == max_profit:
        buy_sell_days.append((dp[-1][i][1], dp[-1][i][2]))
        max_profit -= dp[-1][i][2]
        i -= 1

print("Buy and sell on the following days:")
for day, price in buy_sell_days:
    print(f"Buy on day {day+1} at price {price}")
    print(f"Sell on day {len(prices)} at price {prices[-1]}")
    print()
```
